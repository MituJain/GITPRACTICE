## Variables & types

```python
host = "web-01"        # str
cpu = 87.5              # float
replicas = 3            # int
healthy = True          # bool
nothing = None          # null/None

print(type(host), type(cpu), type(replicas), type(healthy), type(nothing))

# type conversion (very common bug source)
print(int("42"), float("3.14"), str(99), bool(""), bool("x"))
```

## Strings & f-strings (your most-used tool)

```python
name = "web-01"
cpu = 87.456

# f-strings: prefix with f, put expressions in {}
print(f"{name} cpu is {cpu}%")
print(f"{cpu:.1f}")          # 1 decimal place -> 87.5
print(f"{cpu:>10.1f}")       # right-align width 10
print(f"{replicas:03d}")     # zero-pad -> 003  (replicas from cell above)
print(f"{0.873:.0%}")        # percent -> 87%

s = "  ERROR: disk full  "
print(s.strip())             # trim whitespace
print(s.lower(), s.upper())
print(s.strip().replace("ERROR", "WARN"))
print("a,b,c".split(","))    # -> ['a','b','c']
print("-".join(["a", "b", "c"]))  # -> a-b-c
print("disk full".startswith("disk"), "disk full".endswith("full"))
print("error" in "ERROR: disk full".lower())  # substring check
```

## Conditionals

```python
cpu = 92

if cpu > 90:
    status = "critical"
elif cpu > 70:
    status = "warning"
else:
    status = "ok"
print(status)

# ternary / one-liner
alert = "PAGE" if cpu > 90 else "log"
print(alert)

# truthiness: empty things are False
errors = []
if not errors:
    print("no errors")

# chained comparison
print(0 <= cpu <= 100)
```

## Loops

```python
hosts = ["web-01", "web-02", "db-01"]

for h in hosts:
    print(h)

# index + value
for i, h in enumerate(hosts, start=1):
    print(i, h)

# range
for n in range(3):          # 0,1,2
    print(n)

# loop two lists together
loads = [12, 45, 88]
for h, l in zip(hosts, loads):
    print(f"{h}={l}")

# while + break/continue
attempts = 0
while attempts < 5:
    attempts += 1
    if attempts == 2:
        continue   # skip rest of this iteration
    if attempts == 4:
        break      # exit loop
    print("attempt", attempts)
```

## Functions

```python
def health(cpu, mem, threshold=80):   # threshold has a default
    """Return 'bad' if any metric exceeds threshold."""
    if cpu > threshold or mem > threshold:
        return "bad"
    return "good"

print(health(50, 60))               # uses default threshold
print(health(50, 60, threshold=40)) # keyword argument

# *args and **kwargs (variable arguments)
def log(*messages, **fields):
    print("msg:", messages)
    print("fields:", fields)

log("disk full", "on web-01", severity="high", host="web-01")

# return multiple values (tuple)
def minmax(nums):
    return min(nums), max(nums)

lo, hi = minmax([4, 9, 1])
print(lo, hi)

# lambda (tiny anonymous function)
double = lambda x: x * 2
print(double(21))
```

## Comprehensions (Pythonic loops)

```python
loads = [12, 45, 88, 91, 30]

# list comprehension: [expr for item in iterable if condition]
doubled = [x * 2 for x in loads]
high = [x for x in loads if x > 50]
print(doubled)
print(high)

# dict comprehension
hosts = ["web-01", "web-02"]
mapping = {h: 0 for h in hosts}
print(mapping)

# transform + filter together
labels = [f"HIGH:{x}" for x in loads if x > 80]
print(labels)
```

## Error handling (essential for scripts that touch real systems)

```python
def parse_cpu(value):
    try:
        return float(value)
    except ValueError as e:
        print(f"bad value {value!r}: {e}")
        return None
    finally:
        pass  # always runs (cleanup goes here)

print(parse_cpu("87.5"))
print(parse_cpu("N/A"))

# catch multiple exception types
try:
    data = {"cpu": 90}
    print(data["mem"])
except (KeyError, TypeError) as e:
    print(f"missing key: {e}")

# raise your own
def check(cpu):
    if cpu > 100:
        raise ValueError(f"cpu impossible: {cpu}")
    return cpu

try:
    check(150)
except ValueError as e:
    print("caught:", e)
```
