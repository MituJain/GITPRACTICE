# 02 — Data Structures

Lists, dicts, sets, tuples — and the operations you actually use to wrangle metrics, hosts, and log fields.

## Lists (ordered, mutable)

```python
hosts = ["web-01", "web-02", "db-01"]

hosts.append("cache-01")      # add to end
hosts.insert(0, "lb-01")      # add at index
hosts.remove("db-01")          # remove by value
popped = hosts.pop()           # remove & return last
print(hosts, "popped:", popped)

# indexing & slicing
print(hosts[0], hosts[-1])     # first, last
print(hosts[1:3])              # slice [start:stop]
print(hosts[::-1])             # reversed

# useful built-ins
loads = [45, 12, 88, 30]
print(len(loads), sum(loads), min(loads), max(loads))
print(sorted(loads))                       # new sorted list
print(sorted(loads, reverse=True))
print(sorted(hosts, key=len))              # sort by a function
print(88 in loads)                          # membership
```

## Dicts (key-value — the workhorse for JSON/config/labels)

```python
server = {"host": "web-01", "cpu": 87, "region": "us-east"}

print(server["host"])            # KeyError if missing
print(server.get("mem"))         # None if missing (safe)
print(server.get("mem", 0))      # default if missing

server["mem"] = 64               # add / update
del server["region"]             # remove

# iterate
for key, value in server.items():
    print(f"{key} = {value}")

print(list(server.keys()))
print(list(server.values()))
print("cpu" in server)           # key membership

# merge dicts
defaults = {"timeout": 30, "retries": 3}
override = {"retries": 5}
config = {**defaults, **override}   # override wins
print(config)
```

## Counting & grouping (super common in log analysis)

```python
from collections import Counter, defaultdict

status_codes = [200, 200, 404, 500, 200, 404, 500, 500]

# Counter: count occurrences instantly
counts = Counter(status_codes)
print(counts)                       # Counter({200: 3, 500: 3, 404: 2})
print(counts.most_common(2))        # top 2
print(counts[500])                  # count of a specific value

# defaultdict: group without checking if key exists
logs = [("web-01", "error"), ("web-01", "warn"), ("db-01", "error")]
by_host = defaultdict(list)
for host, level in logs:
    by_host[host].append(level)
print(dict(by_host))
```

## Sets (unique values, fast membership)

```python
a = {"web-01", "web-02", "db-01"}
b = {"web-02", "cache-01"}

print(a | b)    # union (all)
print(a & b)    # intersection (in both)
print(a - b)    # difference (in a, not b)
print(a ^ b)    # symmetric diff (in one, not both)

# dedupe a list quickly
dupes = ["web-01", "web-01", "db-01"]
print(set(dupes))
print(len(set(dupes)))  # unique count
```

## Tuples (immutable — fixed records, dict keys)

```python
point = ("us-east", "web-01")
region, host = point          # unpacking
print(region, host)

# tuples can be dict keys (lists cannot)
metrics = {("us-east", "web-01"): 87}
print(metrics[("us-east", "web-01")])

# swap variables in one line
x, y = 1, 2
x, y = y, x
print(x, y)
```

## Nested structures (like real JSON payloads)

```python
cluster = {
    "name": "prod",
    "nodes": [
        {"host": "web-01", "cpu": 87, "tags": ["frontend"]},
        {"host": "db-01", "cpu": 45, "tags": ["data", "primary"]},
    ],
}

print(cluster["nodes"][0]["host"])
print(cluster["nodes"][1]["tags"][1])

# find all hosts over 80% cpu
hot = [n["host"] for n in cluster["nodes"] if n["cpu"] > 80]
print(hot)

# average cpu
cpus = [n["cpu"] for n in cluster["nodes"]]
print("avg cpu:", sum(cpus) / len(cpus))
```
